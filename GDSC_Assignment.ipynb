```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df17.drop(['Label'],axis='columns'),df17.Label,test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)
knn_classifier.fit(x_train, y_train)
knn_classifier.score(x_test,y_test)
```

```python
y_pred = knn_classifier.predict(x_test)
```

```python
specificity_knn3=[]
specificity_knn5=[]
specificity_knn7=[]
sensitivity_knn3=[]
sensitivity_knn5=[]
sensitivity_knn7=[]
```

```python
accuracy_knn3 = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

classes=["BENIGN","DrDoS_LDAP","DrDoS_MSSQL","DrDoS_NTP","DrDoS_NetBIOS","DrDoS_SSDP","DrDoS_UDP","Syn","UDP-lag","WebDDoS"]
print("Accuracy:", accuracy_knn3)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)
total = 0
true = 0
spec=[]
for i in range(len(confusion)):
    for j in range(len(confusion[i])):
        if i==j:
            true += confusion[i][j]
            spec.append(confusion[i][j])
        total+=confusion[i][j]
    print("Sensitivity of class",classes[i],": ",true/total)
    sensitivity_knn3.append(true/total)
print("\n")

for i in range(len(spec)):
    tn=sum(spec)-spec[i]
    k=0
    for j in range(len(confusion[i])):
        if i!=j:
            k+=confusion[i][j]
    print("Specificity of class",classes[i],": ",tn/(tn+k))
    specificity_knn3.append(tn/(tn+k))
```

```python
import matplotlib.pyplot as pyplot
!pip install seaborn
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(confusion,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
